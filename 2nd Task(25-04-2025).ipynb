```python
import pandas as pd

df0 = ['ID','CREDIT_SCORE','VEHICLE_OWNERSHIP',	'MARRIED','CHILDREN','POSTAL_CODE','ANNUAL_MILEAGE','SPEEDING_VIOLATIONS','DUIS','PAST_ACCIDENTS','OUTCOME']
for i in df0:
    print(f"Statistics for column: {i}")
    print(f"Mean: {df[i].mean()}")
    print(f"Median: {df[i].median()}")
    print(f"Standard Deviation: {df[i].std()}")
    print(f"Minimum: {df[i].min()}")
    print(f"Maximum: {df[i].max()}")
    print(f"25th Percentile (Q1): {df[i].quantile(0.25)}")
    print(f"75th Percentile (Q3): {df[i].quantile(0.75)}")
    print(f"Range: {df[i].max() - df[i].min()}")
    print(f"Variance: {df[i].var()}")  
    print("-" * 20)
```

```python
df0 = pd.DataFrame(df, columns=['ID','CREDIT_SCORE','VEHICLE_OWNERSHIP','MARRIED','CHILDREN','POSTAL_CODE','ANNUAL_MILEAGE','SPEEDING_VIOLATIONS','DUIS','PAST_ACCIDENTS','OUTCOME'])

for i in df0.columns:  
    sns.boxplot(x=df0[i])  
    plt.title(f"Boxplot of {i}")
    plt.show()
    print('\n')
```

```python
for i in df0.columns:
    plt.hist(df0[i].dropna(), bins=30, color='skyblue', edgecolor='black')
    plt.title(f'Distribution of {i}')
    plt.xlabel(i)
    plt.ylabel('Frequency')
    plt.grid(True)
    plt.tight_layout()
    plt.show()
```

```python
corr_matrix = df0.corr()

plt.figure(figsize=(10, 7))
sns.heatmap(corr_matrix, annot=True, fmt=".2f", cmap='coolwarm', square=True)
plt.title("Correlation Matrix")
plt.tight_layout()
plt.show()
```

```python
sns.pairplot(df0)
plt.suptitle("Pairplot of Numeric Features", y=1.02)
plt.show()
```

### 1.Correlation Patterns (From Heatmap)

1. Features like INCOME and CREDIT_SCORE show a positive relationship.

2. Age vs Accidents or DUIs might go down as age increases

### 2.Pairplot Patterns
From the pairplot:

1. Clusters of data points, 

2. Outliers that are far away from the rest.

2. Linear or non-linear trends between numeric variables.

**Example:** If AGE vs DUIS shows a trend where younger people have more DUIs, thats a useful pattern.

### 3.Distribution & Anomalies (From Histograms)
From the histogram plots:

1. Is data skewed? (e.g., income might be right-skewed)

3. Are values spread out or clustered tightly?
